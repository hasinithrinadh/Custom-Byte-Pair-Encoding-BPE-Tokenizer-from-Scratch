```python
import numpy as np

tokens = ["low", "est", "</w>"]

print("BPE Tokens:")
print(tokens)

vocabulary = sorted(set(tokens))

print("\nVocabulary:")
print(vocabulary)

token_to_id = {}

for index, token in enumerate(vocabulary):
    token_to_id[token] = index

print("\nToken to ID:")
print(token_to_id)

token_ids = []

for token in tokens:
    token_ids.append(token_to_id[token])

token_ids = np.array(token_ids)

print("\nToken IDs:")
print(token_ids)

embedding_size = 5

embedding_matrix = np.random.rand(len(vocabulary), embedding_size)

print("\nEmbedding Matrix:")
print(embedding_matrix)

embedded_tokens = embedding_matrix[token_ids]

print("\nEmbedded Tokens:")
print(embedded_tokens)

context_vector = np.sum(embedded_tokens, axis=0)

print("\nContext Vector:")
print(context_vector)

scores = context_vector[:len(vocabulary)]

exp_scores = np.exp(scores)

probabilities = exp_scores / np.sum(exp_scores)

print("\nSoftmax Probabilities:")
print(probabilities)
```
